```python
df.columns = df.columns.str.strip()

# Handle null and duplicates
df.fillna(df.mean(numeric_only=True), inplace=True)
df.drop_duplicates(inplace=True)

# Separate features and target
X = df.drop(['Admit', 'Chance of Admit'], axis=1)
y = df['Admit']

# Normalize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


y = y.values.reshape(-1, 1)

```

```python
from scipy.stats import zscore

z_scores = np.abs(zscore(X_scaled))
X_scaled = X_scaled[(z_scores < 3).all(axis=1)]
y = y[(z_scores < 3).all(axis=1)]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic Regression model 
def train_logistic_regression(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    weights = np.zeros((n, 1))
    bias = 0

    for i in range(iterations):
        z = np.dot(X, weights) + bias
        predictions = sigmoid(z)

        
        error = predictions - y
        dw = np.dot(X.T, error) / m
        db = np.sum(error) / m

       
        weights -= learning_rate * dw
        bias -= learning_rate * db

        if i % 100 == 0:
            loss = -np.mean(y * np.log(predictions + 1e-9) + (1 - y) * np.log(1 - predictions + 1e-9))
            print(f"Iteration {i}, Loss: {loss:.4f}")

    return weights, bias

```

```python
def predict(X, weights, bias):
    z = np.dot(X, weights) + bias
    return sigmoid(z) >= 0.5  
```

```python
# Train the model using training data
weights, bias = train_logistic_regression(X_train, y_train, learning_rate=0.1, iterations=1000)
```

```python
y_pred_test = predict(X_test, weights, bias).astype(int)
y_true_test = y_test.astype(int)
```

```python
from sklearn.model_selection import KFold

def manual_cross_val(X, y, k=5):
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    acc_scores = []

    for train_index, val_index in kf.split(X):
        X_train_k, X_val_k = X[train_index], X[val_index]
        y_train_k, y_val_k = y[train_index], y[val_index]

        w, b = train_logistic_regression(X_train_k, y_train_k, learning_rate=0.1, iterations=500)
        y_pred_k = predict(X_val_k, w, b).astype(int)
        acc = accuracy_score(y_val_k, y_pred_k)
        acc_scores.append(acc)

    print("Cross-validation scores:", acc_scores)
    print("Average CV Accuracy:", np.mean(acc_scores))

manual_cross_val(np.array(X_scaled), np.array(y).reshape(-1, 1), k=5)
```

```python
#Exploratory Data Analysis
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report

accuracy = accuracy_score(y_true_test, y_pred_test)
f1 = f1_score(y_true_test, y_pred_test)

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"F1 Score: {f1 * 100:.2f}%")


print("\nClassification Report:")
print(classification_report(y_true_test, y_pred_test))


misclassified = (y_true_test != y_pred_test).sum()
total = len(y_true_test)
print(f"\nMisclassified Samples: {misclassified} out of {total} ({(misclassified / total) * 100:.2f}%)")

cm = confusion_matrix(y_true_test, y_pred_test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
